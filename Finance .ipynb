# **From Dataset Male Are More Likely To Invest Than Female **

```python
# Show frequency age range by gender
df.groupby('gender').age.plot(kind='kde')
plt.xlabel('age')
plt.legend()
plt.show()
```

**The graph show the most people invest between 25-31 years**

```python
import matplotlib.pyplot as plt
df.iloc[:,3:10].sum().plot(kind='bar')
plt.title('The sum of type of investments',fontsize=15)
plt.ylabel('preference(less like)',fontsize=15)
plt.show()
```

**The rank in order of preferrence the data show people prefer to invest public provide fund.**

```python
import seaborn as sns
plt.figure(figsize=(6,6))
sns.countplot(x="Factor",data=df)
plt.title('Factor of Investment')
plt.show()
```

**Often factor considered investing in any instrument is "Return"**

```python
fig, axes = plt.subplots(1, 2,figsize=(15,7))
sns.countplot(ax=axes[0],x=df['Duration'])
axes[0].set_title('Prefer to keep your money')
sns.countplot(ax=axes[1],x=df['Invest_Monitor'])
axes[1].set_title('Invest Monitor')
```

```python
sns.countplot(x=df['Duration'], hue=df['Invest_Monitor'],palette= 'coolwarm' , edgecolor ='black')
plt.legend(loc='upper right')
plt.show()
```

**People usally hold there assets any instruments 1-3 years, 3-5 years and often monitor there investment monthly.**
